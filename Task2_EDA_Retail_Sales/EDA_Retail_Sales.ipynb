```python
import pandas as pd

df = pd.read_csv("menu.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df = df.drop_duplicates()
```

```python
print("Mean Calories:", df["Calories"].mean())
print("Median Calories:", df["Calories"].median())
print("Mode Calories:", df["Calories"].mode()[0])
print("Standard Deviation:", df["Calories"].std())
```

```python
import matplotlib.pyplot as plt

df["Category"].value_counts().plot(kind="bar", figsize=(8,5))
plt.title("Menu Items by Category")
plt.xlabel("Category")
plt.ylabel("Number of Items")
plt.show()
```

```python
import seaborn as sns

sns.histplot(df["Calories"], kde=True)
plt.title("Distribution of Calories")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), cmap="coolwarm")
plt.title("Nutrition Correlation Heatmap")
plt.show()
```
